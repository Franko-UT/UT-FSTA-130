# Q1
   - **Answer**: Simple linear regression is a statistical model that describes the relationship between a dependent variable \( Y \) and an independent variable \( x \). The model is represented by the equation:
     \[
     Y<sub>i</sub> = β<sub>0</sub> + β<sub>1</sub> * x<sub>1</sub> + ϵ<sub>i</sub>
     \]
     where:
     - \( Y<sub>i </sub>\) is the observed value of the dependent variable for the \(i\)th data point.
     - \( x<sub>i </sub>\) is the value of the independent variable for the \(i\)th data point.
     - \( β<sub>0 </sub>\) is the intercept, the predicted value of \( Y \) when \( x = 0 \).
     - \( β<sub>1 </sub>\) is the slope, representing the change in \( Y \) for a one-unit increase in \( x \).
     - \( ϵ<sub>i </sub>\) is the error term for the \(i\)th data point, assumed to be normally distributed with a mean of 0.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the SLR model
beta_0 = 2    # Intercept
beta_1 = 3    # Slope
sigma = 1.5   # Standard deviation of the error term

# Generate predictor variable X
np.random.seed(42)  # For reproducibility
X = np.linspace(0, 10, 100)

# Generate error term epsilon
epsilon = np.random.normal(0, sigma, size=len(X))

# Compute outcome variable Y
Y = beta_0 + beta_1 * X + epsilon

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Sampled Data', alpha=0.7)
plt.plot(X, beta_0 + beta_1 * X, color='red', label='True Regression Line')
plt.title('Sample Data from Simple Linear Regression Model')
plt.xlabel('Predictor Variable X')
plt.ylabel('Outcome Variable Y')
plt.legend()
plt.show()

# Display the distribution of Y for a specific value of X (e.g., X=5)
x_value = 5
y_mean = beta_0 + beta_1 * x_value
y_distribution = norm(loc=y_mean, scale=sigma)

# Plot the distribution of Y at X=5
y_values = np.linspace(y_mean - 3 * sigma, y_mean + 3 * sigma, 100)
plt.plot(y_values, y_distribution.pdf(y_values), label=f'Y | X={x_value}', color='green')
plt.title(f'Normal Distribution of Y at X={x_value}')
plt.xlabel('Y')
plt.ylabel('Density')
plt.legend()
plt.show()
```


# Q2
   - **Answer**: Generate a dataset that follows the specified simple linear regression structure using random samples. For example:

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
     
np.random.seed(42)
x = np.random.rand(100) * 10  # Random independent variable
epsilon = np.random.normal(0, 1, 100)  # Error term
y = 2 + 0.5 * x + epsilon  # Dependent variable based on the model

# Fit the model
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
print(model.summary())
```

# Q4
   - **Answer**: The `fitted_model.fittedvalues` are computed using the equation \(\hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i\), where \(\hat{\beta}_0\) and \(\hat{\beta}_1\) are estimated coefficients found in the summary output's tables.

# Q5
   - **Answer**: The chosen line minimizes the sum of squared residuals, \(\sum (y_i - \hat{y}_i)^2\), providing the best linear fit for the observed data. This approach, known as Ordinary Least Squares (OLS), finds coefficients \(\hat{\beta}_0\) and \(\hat{\beta}_1\) that achieve this minimization.

# Q6
   - **Answer**: The expression for \( R^2 \) is:
     \[
     R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
     \]
     It measures how much of the total variability in \( Y \) is explained by the fitted model. An \( R^2 \) close to 1 indicates a good fit.

     - **Independence**: Observations are independent of each other.
     - **Homoscedasticity**: The variance of the errors \(\epsilon_i\) is constant.
     - **Normality**: The error terms \(\epsilon_i\) are normally distributed.

   - **Answer**: The null hypothesis \( H_0 \) states:
     \[
     H_0: \beta_1 = 0
     \]
     This hypothesis indicates that there is no linear association between \( Y \) and \( x \).

# Q9
   - **Answer**: The vertical spread of the \( Y_i \) values around the line of best fit affects the strength of the evidence against the null hypothesis. If the data points are tightly clustered around the line, it supports rejecting \( H_0 \). A wide spread indicates weaker evidence against \( H_0 \).
